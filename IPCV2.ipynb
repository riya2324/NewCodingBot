```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
image = cv2.imread(list(uploaded.keys())[0], cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis('off')
```

```python
negative_image = 255 - image

plt.imshow(negative_image, cmap='gray')
plt.title("Image Negative")
plt.axis('off')
```

```python
r_min = np.min(image)
r_max = np.max(image)

contrast_stretched = ((image - r_min) / (r_max - r_min)) * 255
contrast_stretched = contrast_stretched.astype(np.uint8)

plt.imshow(contrast_stretched, cmap='gray')
plt.title("Contrast Stretched Image")
plt.axis('off')
```

```python
plt.figure(figsize=(10,6))

for i in range(8):
    bit_plane = (image >> i) & 1
    plt.subplot(2,4,i+1)
    plt.imshow(bit_plane, cmap='gray')
    plt.title(f'Bit Plane {i}')
    plt.axis('off')

plt.tight_layout()
```

```python
lower = 100
upper = 180

gray_slice = np.zeros_like(image)
gray_slice[(image >= lower) & (image <= upper)] = 255

plt.imshow(gray_slice, cmap='gray')
plt.title("Gray Level Slicing (Without Background)")
plt.axis('off')
```

```python
gray_slice_bg = image.copy()
gray_slice_bg[(image >= lower) & (image <= upper)] = 255

plt.imshow(gray_slice_bg, cmap='gray')
plt.title("Gray Level Slicing (With Background)")
plt.axis('off')
```
