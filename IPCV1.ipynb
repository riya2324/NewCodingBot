```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
image_path = '/content/drive/My Drive/IPCV(LAB-1)/Sample_CV.webp'
```

```python
img = cv2.imread(image_path)
```

```python
if img is None:
    print(f"Error: Could not load image from {image_path}. Please verify the file path, name, extension, and ensure the file is not corrupted.")
else:
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.title("Original Image (RGB)")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(img_gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis("off")

plt.tight_layout()
plt.show()
```
